- strip(), split() 함수 사용
- #encoding='cp949' 또는 #encoding='utf-8'

```python
f = open('인물.txt',encoding='utf-8')
s = f.readline()
print(s)

data = [] 

for line in f:
    l = line.strip().split(',') # 2차원 리스트로 변환, 문자열로 기본
    l[1] = int(l[1]) # 숫자 형식(정수)으로 바꾸기
    data.append(l)

f.close()
print(data)
```

## 클래스
- 클래스는 청사진을 만드는 것이다
- 설계를 잘 해야 클래스를 잘 만들 수 있다
- 클래스가 현실화 된 것이 *객체(Object)*이다

```python
class ABC:
    def __init__(self):
        self.name='ABC' # 변수 선언
        self.count=3 # 변수 선언
        
    def first(self): # 함수 정의시 첫번째 인자는 반드시 self
        print('I am from', self.name)
    
    def second(self, n):
        return n+self.count
    
    def third(self):
        self.first()
        print('Test: ')

    
abc = ABC() # 객체 생성
abc.first() # 함수를 호출할 때는 self를 쓰지 않는다
print(abc.second(2))
print(abc.name, abc.count)
print(abc.third())
```
