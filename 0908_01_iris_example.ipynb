### 연습문제
======

iris 데이터를 불러오자.

1. iris 의 50번째~99번째 까지의 샘플에서
두번째,세번째,다섯번째 속성을 뽑아내시오.
(2차원 리스트로 생성)

2. 각 속성의 평균을 구하고,
각 속성에서 해당 평균을 뺀 결과 리스트를 구하시오.

* 깃허브의 어제 강의노트 참고


```python
f = open('iris.csv')

s = f.readline()
header = s.strip().split(',')

labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
iris = []

for line in f:
    l = line.strip().split(',')
    l[:4] = list(map(float, l[:4]))
    l[4] = labels.index(l[4])
    iris.append(l)
    
print(iris)
f.close()
```

```python
[[i[1],i[2],i[-1]] for i in iris[50:100]]
```

```python
avg = [0,0,0,0]

for i in iris:
    for j in range(4):
        avg[j] += i[j]
avg = [i/len(iris) for i in avg]
avg
```

```python
# 평균
a = [0,0,0,0]

for i in iris:
    for j in range(4):
        a[j] += i[j]
a = [i/len(iris) for i in a]

# 빼기
iris2 = iris.copy()

for i in iris2:
    for j in range(4):
        i[j] -= a[j]
iris2
    
```

## 머신러닝
- 가까운 이웃 알고리즘을 이용한 분류

```python
from sklearn.neighbors import KNeighborsClassifier

x = iris[:, :4] # 데이터
y = iris[:, 4] # 목표값

knn = KNeighborsClassifier() # 머신(엔진) 생성
knn.fit(x,y) # 훈련한다
knn.score(x,y) # 점수(평가)
```

```python
knn.predict(x)
```

- 근사하는 직선 긋기 (회귀)

```python
from sklearn.linear_model import LinearRegression

linear = LinearRegression()
linear.fit(iris[:,[2]], iris[:,3])
```

```python
기울기 = linear.coef_[0]
절편 = linear.intercept_
기울기, 절편
```

```python
plt.scatter(iris[:,2], iris[:,3])
plt.xlabel('PetalLength')
plt.ylabel('PetalWidth')

plt.plot([0,7], [절편, 7*기울기+절편], 'r:')
```
