### 연습문제
- 아이리스 데이터에서, 첫번째 속성(sepal length)의 값이 평균값 보다 큰 샘플들을 뽑아내시오.
- 문제1) 파이썬의 일반적인 방법으로 풀어보시오.
- 문제2) 일반적인 방법 이외에, 먼저 해당 인덱스를 먼저 알아낸 다음 팬시 색인을 적용해보시오.

```python
iris_labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
iris = np.loadtxt('iris.csv', skiprows=1, delimiter=',',
                  converters={4: lambda s: iris_labels.index(s.decode())})
iris
```

```python
avg = [0,0,0,0]

for i in iris:
    for j in range(4):
        avg[j] += i[j]
avg = [i/len(iris) for i in avg]
avg
```

```python
idx = []
s = iris[:,0].mean()

for i, l in enumerate(iris): # idx를 뽑아내기 위해 일단 행 번호를 알아야함 => enumerate or range()
    if l[0] > s:
        idx.append(i)
        
display(iris[idx], idx,len(idx))
```

```python
# 문제1

data =[]

m = iris[:,0].mean()

for l in iris:
    if l[0] > m:
        data.append(l)
data = np.array(data)
display(m, data, data.shape)
```

```python
# 문제2
idx = []
m = iris[:,0].mean()

for i in range(150):
    if iris[i,0] > m:
        idx.append(i)
display(iris[idx], idx,len(idx))
```
