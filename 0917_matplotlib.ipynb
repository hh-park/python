### Weight(가중치)
- 값들에 가중치를 곱하는 것 -> 신경망의 기본

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
score = np.random.randint(50,100,size=[100,3])
W=np.array([1,0.5,0.7])
score
```

```python
(score*W).mean(axis=1) # 가중치 곱하기
```

```python
(score@W)/3 # 가중치 곱하기 (*)
```
