### Numpy array - iris 데이터 읽어오기(1)

```python
f = open('iris.csv')

s = f.readline()
header = s.strip().split(',')

labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
data = []

for line in f:
    l = line.strip().split(',')
    l = [float(l[0]), float(l[1]), float(l[2]), float(l[3]), labels.index(l[4])]
    data.append(l)

f.close()
```

```python
# Numpy array 로 변환한다
iris = np.array(data)
iris
```

### arange() 함수
- reshape()
- numpy는 최대한 원본 데이터를 재사용하도록 설계되었다

```python
a = np.arange(100)
a
```

```python
a.shape
```

```python
a.reshape(10,10)
```

```python
np.arange(100).reshape(5,20)
```

```python
a=np.arange(100)
b=a.reshape(10,10) # b는 a를 가리키고 있기 때문에 원본이 같음
b[0,0] = 9999
a # 원본도 바뀐다
# 그럴지라도 대용량 데이터는 copy를 잘 안함, 
# numpy는 되도록이면 '참조', 원본을 그래도 활용한다
```

```python
# list는 slice 하면 copy 된다
l = [1,2,3,4,5]
l2 = l[1:4]
l2
```

```python
l2[0] = -1
l
```

```python
# numpy의 array는 copy가 아니라 참조한다 => 그래서 속도가 빠르다
a = np.array([1,2,3,4,5])
a2 = a[1:4]
a2
```

```python
a2[0] =-1
a
```

```python
np.arange(100).reshape(2,5,10)
```

```python
np.arange(100).reshape(10,-1)
```

```python
a = np.arange(100).reshape(10,10)
b = a.T # transpose (전치행렬)
b
```

```python
import matplotlib.pyplot as plt
plt.imshow(a) # a.T
plt.colorbar()
```

```python
import matplotlib.pyplot as plt
a = np.arange(100).reshape(10,10)
plt.imshow(a, cmap='gray') #gray_r
plt.colorbar()
```

### 연습문제
- 10x10 형태의 array를 만들고, 테두리만 까맣게 그리시오(plt.imshow()사용)
- 체스판 무늬를 그리시오(값은 0과 1을 사용한다)

```python
a = np.array(([0,1]*4 + [1,0]*4)*4).reshape(8,8)
a
plt.imshow(a, cmap='gray_r')
plt.colorbar()
```

```python
b = np.arange(100).reshape(10,10)
b[:,:] = 1
b[1:9, 1:9] = 0
plt.imshow(b, cmap='gray_r')
plt.colorbar()
```
