### 연습문제
- 동전을 100번 던지는 것을 한 게임이라고 하자.
> 앞면이 나오면 이기는 것이다.
- 이런 게임을 10000번 했을 때, 이를 numpy 어레이로 표현해 보자.
> np.random.randint() 함수 사용
- 각 게임에서 이긴 횟수를 기록하고, 그 결과를 아래와 같은 리스트로 나타내자.
> [0번 이긴 게임수, 1번 이긴 게임수, ... , 100번 이긴 게임수]
- 위의 결과를 plt.plot() 함수로 그려보자.
> 종모양 곡선이 나오는지 확인해 보시오.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
match = np.random.randint(2, size = [10000,100]) # 100번 던진 걸(game) 만 번(match) 했다
match
```

```python
wins = [game.sum() for game in match]
wins
```

```python
counts = np.zeros(101)
for i in wins:
    counts[i] += 1 # 이긴 게임수에 해당하는 카운트를 증가시킴

counts

# counts = np.bincount(wins)
```

```python
plt.plot(counts) # 종모양 => 큰 수의 법칙
```

```python
sum(wins)/len(wins)
```

> 가위바위보

```python
np.random.seed(200)
g = np.random.randint(3, size =10)
```

```python
t = np.random.randint(3, size =10)
result = np.zeros(10)

for i in range(9):
    if t[i] > g[i]:
        result[i] +=1
        
    else: result[i] -=1
        
display(g,t,result)
```

### 이미지 출력
- 읽어오기 : plt.imread()
- 그리기 : plt.imshow()
- 흑백이미지는 2차원 => (높이, 넓이)
- 컬러이미지는 3차원 => (높이, 넓이, 채널)
- Gray scale => min: 0(black), max: 255(white)

```python
# 이미지 출력
a = plt.imread('home.png')
print(a.shape, a.ndim)

plt.imshow(a)
plt.title('home')
```

```python
a = plt.imread('1.jpg')
print(a.shape, a.ndim)

plt.imshow(a)
plt.title('home')

plt.figure(figsize=[10,4])

plt.subplot(2,3,1)
plt.imshow(a[:,:,0], cmap='gray', vmin=0, vmax=255)

plt.subplot(2,3,2)
plt.imshow(a[:,:,1], cmap='gray', vmin=0, vmax=255)

plt.subplot(2,3,3)
plt.imshow(a[:,:,2], cmap='gray', vmin=0, vmax=255)

plt.subplot(2,3,4)
plt.imshow(a)
```

```python
b = plt.imread('snoopy.jpg')
print(b.shape, b.ndim)

plt.imshow(b)
plt.title('Snoopy')

plt.figure(figsize=[10,4])

plt.subplot(2,3,1)
plt.title('red') # 빨강이 제일 많아서 제일 밝음
plt.imshow(b[:,:,0], cmap='gray', vmin=0, vmax=255)
plt.subplot(2,3,2)
plt.title('green')
plt.imshow(b[:,:,1], cmap='gray', vmin=0, vmax=255)
plt.subplot(2,3,3)
plt.title('blue') # 파랑 부분이 제일 밝게 됨
plt.imshow(b[:,:,2], cmap='gray', vmin=0, vmax=255)
plt.subplot(2,3,4)
plt.imshow(b)
```

```python
b1 = b.copy()
b1[200:400,400:1000,2] = 0 # 2 - 0: blue를 0으로 놨다
plt.imshow(b1)
```

```python
# plt.imshow(b1[:300, :300]) # 잘라내기
# plt.imshow(255-b) # 네거티브 이미지
b1 = b.copy()
b1[:,:,2] = 0 # 파란색 제거
plt.imshow(b1)
```
