#### 연습문제 (~4시40분)
- 아이리스 데이터를 X, y 로 분리하시오.
- X 에서 petal width 값이 2보다 큰 샘플들의 행번호를 구하시오.
- 그 행번호를 가지고 y 값을 출력하시오.
- plt.plot() 함수를 사용하여 별모양을 그리시오.

```python
X = np.loadtxt('iris.csv', delimiter=',', skiprows=1, usecols=range(4))
y = np.array([0]*50+[1]*50+[2]*50)
```

```python
np.where(X[:,-1]>2)
```

```python
idx = np.where(X[:,-1]>2)[0]
idx
```

```python
y[idx]
```

```python
plt.scatter(X[idx,0], X[idx,1], c='r', s=200, marker='*')
plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
```

```python
#plt.scatter(X[idx,0], X[idx,1], c='r', s=200, marker='*')
plt.scatter(X[:,0], X[:,1], c=y, s=np.exp(X[:,-1]**2), alpha=0.5)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
```
